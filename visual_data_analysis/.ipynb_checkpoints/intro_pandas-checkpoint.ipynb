# Introduction to Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# an easy way to run magic commands, like ls on current dir
! ls
```

```python
# read excel data file
file = "gapminder_population.xlsx"

xls = pd.ExcelFile(file)

print(xls.sheet_names)
```

```python
# parse data spreadsheet into a pd dataframe
df = xls.parse('Data')

df.head()
```

```python
# data size
df.shape
```

```python
# there is an open bug in pandas when running info method for df with integer column names that are not starting 0 
# consecutive https://github.com/pandas-dev/pandas/issues/37408
#df.info()
```

```python
df.columns.map(type)
```

```python
# changing column names to str

df.columns = df.columns.astype(str)

df.columns.map(type)
```

```python
df.info()
```

```python
# get summary statistics of numerical columns
df.describe()
```

```python
# ordering rows 

df.sort_values('Total population', ascending=False, inplace=True)

df.head()
```

```python
# resetting index
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
# rename column
df.rename(columns = {'Total population':'country'}, inplace = True)
```

```python
# slicing multiple columns
# get only data for 1800
columns_subset = ['country']+[col for col in df.columns if col != 'country' and 
                                       int(col)>=1800 and int(col)<1900]
df[columns_subset].head()
```

```python
#filtering rows 
zambia = df[df['country'] == 'Zambia']
zambia
```

```python
#getting values from dataframe
zambia.values
```

```python
# using .isin to filter
zcountries = df[df['country'].isin(['Zambia', 'Zimbabwe'])]

zcountries
```

```python
# indexing
# get population of Yemen in 1995

yemen_1995 = df[df['country']=='Yemen']['1995']

yemen_1995.values[0]
```

```python
# to make things easier to index, we can set country as index

df.set_index('country', inplace=True)
df.head()
```

```python
# indexing using loc
yemen_1995 = df.loc['Yemen', '1995']
yemen_1995
```

```python
# another way to filter rows(axis=0) and columns(axis=1), is to use filter
# filter data for scandinavian countries

scandi_countries = ['Sweden', 'Finland', 'Denmark', 'Norway']

scandi = df.filter(scandi_countries, axis=0)

scandi.head()
```

```python
# we can use various summary stats on column(s)

year2000_stats = df['2000'].describe()

year2000_stats
```

```python
# we can also define a custom aggregate function

def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

year200_iqr = df['2000'].agg(iqr)

year200_iqr
```
