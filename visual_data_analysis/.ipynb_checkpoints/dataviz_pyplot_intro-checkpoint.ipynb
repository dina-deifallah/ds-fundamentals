# Introduction to Data Viz with matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
! ls
```

```python
diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

```python
flights = sns.load_dataset("flights")

flights.info()
```

```python
flights.describe()
```

## 1. Syntax for creating a figure


```python
fig, ax = plt.subplots()
plt.show()
```

## 2. Adding data to axes of a figure

```python
fig, ax = plt.subplots()

flights_1950 = flights[flights['year']==1950]
flights_1960 = flights[flights['year']==1960]

ax.plot(flights_1950['month'], flights_1950['passengers'])
ax.plot(flights_1960['month'], flights_1960['passengers'])


plt.show()
```

## 3. Configuring a plot

```python
fig, ax = plt.subplots(figsize=(8,6))

flights_1950 = flights[flights['year']==1950]
flights_1960 = flights[flights['year']==1960]

ax.plot(flights_1950['month'], flights_1950['passengers'], color='r', linestyle='--', marker='v', label='1950')
ax.plot(flights_1960['month'], flights_1960['passengers'], color='b', linestyle='--', marker='o', label='1960')

ax.set_xlabel('time (months)')
ax.set_ylabel('total number of passengers')
ax.set_title('Comparing Flight Passengers between years 1950 and 1960')
ax.legend()
plt.xticks(rotation=45)
plt.show()
```

## 4. Subplots

```python
fig, ax = plt.subplots(2,1, figsize=(8,8), sharey=True)

ax[0].plot(flights_1950['month'], flights_1950['passengers'], color='r', linestyle='--', marker='v', label='1950')
ax[1].plot(flights_1960['month'], flights_1960['passengers'], color='b', linestyle='--', marker='o', label='1960')

ax[0].set_ylabel('total number of passengers(1950)')
ax[1].set_ylabel('total number of passengers(1960)')
ax[1].set_xlabel('time (months)')

plt.legend()
plt.show()
```

## 5. Visualizing time series data

```python
climate = pd.read_csv("DailyDelhiClimateTrain.csv", parse_dates=['date'])

climate.head()
```

```python
print(climate.info())
print()
print(climate.describe())
```

```python
# to make subsetting data by date easier, use the date column as the index for the dataframe
climate.set_index('date', inplace=True)
climate.head()
```

```python
print(climate.index.min(), climate.index.max())
```

```python
climate_2013 = climate["2013-01-01": "2013-12-31"]

climate_2014 = climate["2014-01-01": "2014-12-31"]

print(climate_2013.index.min(), climate_2013.index.max())
print(climate_2014.index.min(), climate_2014.index.max())
```

```python
# show temp trend over the given data

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(climate.index, climate['meantemp'], linestyle='--', marker='o')
plt.xticks(rotation=45)
plt.show()
```

```python
# plot tempreture and wind speed in 2013

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(climate_2013.index, climate_2013['wind_speed'], color='r', linestyle='--')
ax.set_ylabel('Wind Speed (km/hr)', color='r')
ax.tick_params('y', colors='r')
ax.set_xlabel('Time')

# create a twin axes that shares the x-axis
ax2 = ax.twinx()
ax2.plot(climate_2013.index, climate_2013['meantemp'], color='b', linestyle='--')
ax2.set_ylabel('Temprature (Celsius)', color='b')
ax2.tick_params('y', colors='b')

ax.set_title('Wind Speed and average temprature in India in 2013')
plt.xticks(rotation=45)
plt.show()
```

```python
# Compare average temprature between 2013 and 2014

fig, ax = plt.subplots(figsize=(10,6))

ax.plot(climate_2013.index, climate_2013['meantemp'], color='r', linestyle='--')
ax.set_ylabel('Temprature (Celsius)')
ax.tick_params('x', colors='r')
plt.xticks(rotation=45)
ax.set_xlabel('Time')

# create a twin axes that shares the y-axis
ax2 = ax.twiny()
ax2.plot(climate_2014.index, climate_2014['meantemp'], color='b', linestyle='--')
ax2.tick_params('x', colors='b')

ax.set_title('Average temprature in India in 2013 and 2014')
plt.xticks(rotation=45)
plt.show()
```

## 6. Comparative plots

### 6.1 Bar plots

```python
# plot the number of diamonds in the diamond data set for each color that have an ideal, premium and very good cuts
# using a stacked bar chart
diamonds_cc = pd.crosstab(diamonds['color'], diamonds['cut'])
diamonds_cc
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.bar(diamonds_cc.index, diamonds_cc['Ideal'], label='ideal')
ax.bar(diamonds_cc.index, diamonds_cc['Premium'], label='premium', bottom=diamonds_cc['Ideal'])
ax.bar(diamonds_cc.index, diamonds_cc['Very Good'], 
       label='very good', bottom=diamonds_cc['Ideal']+diamonds_cc['Premium'])
ax.set_xlabel("diamonds color")
ax.set_ylabel("diamonds number")
plt.legend()
plt.show()
```

### 6.2 Histograms

```python
# compare between the distributions of mean temp in india in 2013 and 2014

fig, ax = plt.subplots(figsize=(10,8))

ax.hist(climate_2013['meantemp'], label='2013', bins=20, alpha=0.5) # default number of bins is 10
ax.hist(climate_2014['meantemp'], label='2014', bins=20, alpha=0.5)
ax.set_xlabel("Mean Temperature (Celsius)")
plt.legend()
plt.show()
```

### 6.3 Boxplots

```python
# compare between the distributions of windspeeds in india in 2013 and 2014. Are there any outliers ?

fig, ax = plt.subplots(figsize=(8,6))
ax.boxplot([climate_2013['wind_speed'], climate_2014['wind_speed']])
ax.set_xticklabels(["2013", "2014"])
ax.set_ylabel('Wind Speed (km/hr)')

plt.show()
```

### 6.4 Scatter plots

```python
# compare between humidity and mean temprature in india over time, is there any obvious correlation ?

fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(climate['humidity'], climate['meantemp'], c=climate.index)
ax.set_xlabel('Humidity')
ax.set_ylabel('Mean Temprature (Celcius)')
plt.show()
```
