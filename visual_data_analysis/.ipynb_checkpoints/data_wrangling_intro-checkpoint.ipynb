```python
# import multiple files using grob and wildcards

filenames = glob('DailyDelhiClimate*.csv')

dataframes = [pd.read_csv(f, parse_dates=True) for f in filenames]
```

```python
ClimateTest = dataframes[0]
ClimateTrain = dataframes[1]

print(ClimateTest.shape, ClimateTrain.shape)
```

```python
ClimateTrain.head()
```

## 1. Sorting Index

```python
ClimateTrain.set_index('date', inplace=True, drop=True)
```

```python
ClimateTrain.head()
```

```python
# sorting index descendingly

ClimateTrain.sort_index(ascending=False, inplace=True)
ClimateTrain.head()
```

```python
# sort df using the values of a column

ClimateTrain.sort_values("meantemp", ascending=False, inplace=True)
ClimateTrain.head(10)
```

## 2. Arithmetic operations

```python
ClimateTrain.reset_index(inplace=True)
ClimateTrain['date'] = pd.to_datetime(ClimateTrain['date'])
ClimateTrain.set_index('date', inplace=True, drop=True)
```

```python
# calculate the monthly change in the average meantemp from 2013 to 2015

ClimateTrain.sort_index(inplace=True)
temperature = ClimateTrain.loc['2013':'2015', ['meantemp']]
temperature.head()
```

```python
# resample to get the monthly mean
monthly = temperature.resample('M').mean()
```

```python
monthly.head()
```

```python
# calculate percentage change 
monthly['percent_change']= monthly.pct_change()*100
```

```python
monthly.head()
```

## 3. Concatenating dataframes

```python
ClimateTest['date'] = pd.to_datetime(ClimateTest['date'])
ClimateTest.set_index('date', inplace=True)
print(ClimateTrain.head())
print(ClimateTest.head())
```

```python
# combine data from both dataframes into one dataframe

Climate = pd.concat([ClimateTrain, ClimateTest], axis='rows')
Climate.sort_index(inplace=True)

print(Climate.head())
print(Climate.tail())
```

## 4. Merging dataframes

```python
data_url = "https://raw.githubusercontent.com/wblakecannon/DataCamp/master/14-merging-dataframes-with-pandas/_datasets/summer-olympic-medals/Summer%20Olympic%20medalists%201896%20to%202008%20-%20EDITIONS.tsv"
olympics = pd.read_csv(data_url, sep='\t')

olympics.head()
```

```python
olympics.info()
```

```python
data_url2 = "https://raw.githubusercontent.com/johnashu/datacamp/master/medals/Summer%20Olympic%20medalists%201896%20to%202008%20-%20IOC%20COUNTRY%20CODES.csv"
codes = pd.read_csv(data_url2)

codes.head()
```

```python
codes.info()
```

```python
data_url3 = "https://raw.githubusercontent.com/johnashu/datacamp/master/medals/Summer%20Olympic%20medalists%201896%20to%202008%20-%20ALL%20MEDALISTS.tsv"
medals = pd.read_csv(data_url3, sep='\t', skiprows=4)
medals.head()
```

```python
medals.info()
```

```python
#count medals by country and edition

medals_count = medals.pivot_table(index='Edition', values='Athlete', columns='NOC', aggfunc='count')
medals_count.head()
```

```python
# convert counts to percentage of total medals, total medals each edition stored in olympics

total_medals = olympics.set_index('Edition')
total_medals = total_medals['Grand Total']
total_medals.head()
```

```python
medals_perc = medals_count.divide(total_medals, axis='rows')*100

medals_perc.head()
```

```python
# calculate the running mean, i.e. instead of actual values, show mean up until this row
mean_medals_perc = medals_perc.expanding().mean()

mean_medals_perc.head()
```

```python
# calculate running percentage changes
change_medals_perc = mean_medals_perc.pct_change()*100
change_medals_perc.head()
```

```python
# construct a dataframe of NOC of host countries each edition
hosts = pd.merge(olympics, codes, how='left', on='Country')

print(hosts.head())
print(hosts.info())
```

```python
# fix missing values
print(hosts.loc[hosts.NOC.isnull()])
```

```python
hosts.set_index('Edition', inplace=True)
hosts.loc[1972, 'NOC'] = 'FRG'
hosts.loc[1980, 'NOC'] = 'URS'
hosts.loc[1988, 'NOC'] = 'KOR'
```

```python
hosts.reset_index(inplace=True)
print(hosts.loc[hosts.NOC.isnull()])
```

```python
# transform change in running mean of medals percentage per country dataframe from wide to long format
change_medals_perc.reset_index(inplace=True)
reshaped = pd.melt(change_medals_perc, id_vars='Edition', value_name='Change')
reshaped.loc[reshaped.NOC=='USA']
```

```python
# check if percentage change is positive for countries when they are hosts

# Merge reshaped and hosts
merged = pd.merge(reshaped, hosts)
print(merged.head())

# Set Index of merged and sort it
influence = merged[['Edition', 'NOC', 'Country', 'Change']].set_index('Edition').sort_index()

print(influence.head())
```

```python
# plot influence as a bar plot

fig, ax = plt.subplots(figsize=(15,8))
ax.bar(influence['Country'], influence['Change'])
ax.set_ylabel("Percentage change in number of olympic medals when hosting")
plt.xticks(rotation=90)
plt.show()
```
