### 1.2 Making recommendations based on frequency of pairs of movies seen together by the same user

```python
# define a custom function that returns permulations of movies 

def find_movie_pairs(x):
    ''' creates a pandas Dataframe containing all possible pairs of items from a pandas series'''
    
    pairs_list = list(permutations(x.values, 2))
    pairs_df = pd.DataFrame(pairs_list, columns=['movie_a', 'movie_b'])
    
    return pairs_df

```

```python
# Apply the function to the title column and reset the index

movie_pairs = movies.groupby('userId')['title'].apply(find_movie_pairs).reset_index(drop=True)
```

```python
movie_pairs.head()
```

```python
movie_combinations = movie_pairs.groupby(['movie_a', 'movie_b']).size()

movie_combinations
```

```python
# convert nested series to a dataframe and reset index

movie_combinations = movie_combinations.to_frame(name='count').reset_index()

movie_combinations.head()
```

```python
# Sort the counts from highest to lowest

movie_combinations.sort_values(by='count', ascending=False, inplace=True)

movie_combinations.head()
```

```python
# Find the movies most frequently watched by people who watched Thor

thor_df = movie_combinations[movie_combinations['movie_a']=='Thor (2011)'].reset_index()
```

## 2. Item-based recommendations

### 2.1 Jaccard similarity

```python
movies.head()
```

```python
movies.info()
```

```python
# splitting year from title

movies[['title_', 'year']] = movies['title'].str.split('(', 1, expand=True)

movies['year'] = movies['year'].str.replace(')', '')

movies['title_'] = movies['title_'].str.strip()

movies.head()
```

```python
# number of unique movies we have

print("number of movies in dataset = {}".format(movies['title'].nunique()))
```

```python
# create a movie genre dataframe with columns movie name and genres


movie_genre = movies[['title', 'title_', 'genres']].drop_duplicates(subset='title', ignore_index=True)

movie_genre.shape
```

```python
movie_genre.set_index('title', inplace=True)
```

```python
# convert genre column string values to a list

movie_genre_series = movie_genre['genres'].str.split('|')


movie_genre_df = movie_genre_series.to_frame(name='genre_list').reset_index()


movie_genre_df.head()
```

```python
# convert dataframe to long format

movie_genre_longdf = movie_genre_df.explode(column='genre_list')

movie_genre_longdf.head()
```

```python
# use crosstab to create version we need

movie_genre_ct = pd.crosstab(movie_genre_longdf['title'], movie_genre_longdf['genre_list'])

movie_genre_ct.head()
```

```python
movie_genre_ct.describe()
```

```python
# using the jaccard score to measure the similarity between Toy Story and Thor

toy_story_array = movie_genre_ct.loc['Toy Story (1995)'].values

thor_array = movie_genre_ct.loc['Thor (2011)'].values

iron_man_array = movie_genre_ct.loc['Iron Man (2008)'].values

avengers_array = movie_genre_ct.loc['Avengers, The (2012)'].values



print(jaccard_score(toy_story_array, thor_array))
print(jaccard_score(iron_man_array, thor_array))
print(jaccard_score(iron_man_array, avengers_array))
```

```python
# calculating pairwise jaccard similarity score using pdist from scipy (similarity = 1 - distance)

jaccard_distances = pdist(movie_genre_ct, metric='jaccard')

jaccard_distances
```

```python
# arranging in square format

jaccard_similarity_array = 1 - squareform(jaccard_distances)

jaccard_similarity_df = pd.DataFrame(jaccard_similarity_array, 
                                     columns=movie_genre_ct.index, index=movie_genre_ct.index)
```

```python
jaccard_similarity_df.head()
```

```python
# getting recommedations using jaccard similarity for Thor

Thor_series = jaccard_similarity_df.loc['Thor (2011)'].sort_values(ascending=False)
Thor_series[1:11] # top 10 recommendations 
```

```python
# wrapping it up in a function

def get_recommendations_by_movie(movie_name, similarity_df):
    """
    returns the top 10 recommended movies for a given movie name based on Jaccard similarity score
    
    input:
    movie_name (str)
    similarity_df (data frame)
    
    returns:
    list of 10 most similar movies 
    """
    
    recommendations = similarity_df.loc[movie_name].sort_values(ascending=False)
    
    return list(recommendations[1:11].index)
```

```python
get_recommendations_by_movie('Thor (2011)', jaccard_similarity_df)
```

### 2.3 Creating recommendations based on user profile

```python
movies.head()
```

```python
def get_user_profile(user_id, movies_df):
    
    """
    function to generate list of watched movies by a given user id that she/he gave at least 4 as a rating
    
    """
    
    user_profile = movies_df.loc[(movies_df['userId']==user_id) & (movies_df['rating'] >= 4)]['title'].values
    
    return list(user_profile)
```

```python
def get_user_full_profile(user_id, movies_df):
    """
    function to generate the full list of watched movies by a given user id
    
    """
    user_full_profile = movies_df.loc[movies_df['userId']==user_id]['title'].values
    
    return list(user_full_profile)
```

```python
user_profile_1 = get_user_profile(1, movies)

user_profile_1
```

```python
# generating user profile dataframe

user_profile_df = tfidf_df.reindex(user_profile_1)

user_profile_df.head()
```

```python
user_profile_summary = user_profile_df.mean()

user_profile_summary
```

```python
# finding all movies the user did not watch

user_full_profile = get_user_full_profile(1, movies)


tfidf_df_subset = tfidf_df.drop(user_full_profile, axis=0)

tfidf_df_subset.head()
```

```python
# applying cosine similarity between user profile and movies not yet watched

user_cosine_similarity_array = cosine_similarity(user_profile_summary.values.reshape(1,-1), tfidf_df_subset)

user_cosine_similarity_array.shape
```

```python
# creating a dataframe and sorting it to generate personalized recommendations

user_cosine_similarity_df = pd.DataFrame(user_cosine_similarity_array.T, columns=['similarity'], 
                                         index=tfidf_df_subset.index)

user_cosine_similarity_df.sort_values(by='similarity', ascending=False, inplace=True)

user_cosine_similarity_df.head(100)
```

### 4.2 Singular Value Decomposition

```python
# Get the average rating for each user 
avg_ratings = user_ratings_table.mean(axis=1)

# Center each user's ratings around 0
user_ratings_centered = user_ratings_table.sub(avg_ratings, axis=0)

# Fill in all missing values with 0s
user_ratings_centered.fillna(0, inplace=True)

# Print the mean of each column
print(user_ratings_centered.mean(axis=1))
```

###  we will decompose the centered user ratings matrix into 3 factors: U, sigma, and Vt.

1. U is a matrix with a row for each user
2. Vt has a column for each movie
3. sigma is an array of weights that you will need to convert to a diagonal matrix


```python
from scipy.sparse.linalg import svds
import numpy as np
```

```python
# Decompose the matrix
U, sigma, Vt = svds(user_ratings_centered)
```

```python
# Convert sigma into a diagonal matrix
sigma = np.diag(sigma)
print(sigma)
```

```python
# multiplying the three matrices to get the filled user ratings matrix

# Dot product of U and sigma
U_sigma = np.dot(U, sigma)

# Dot product of result and Vt
U_sigma_Vt = np.dot(U_sigma, Vt)

# Add back on the row means contained in avg_ratings
uncentered_ratings = U_sigma_Vt + avg_ratings.values.reshape(-1, 1)

# Create DataFrame of the results
calc_pred_ratings_df = pd.DataFrame(uncentered_ratings, 
                                    index=user_ratings_table.index,
                                    columns=user_ratings_table.columns
                                   )
# Print the recalculated matrix 
calc_pred_ratings_df.head()
```

```python
# Sort the ratings of User 5 from high to low
user_5_ratings = calc_pred_ratings_df.loc[5,:].sort_values(ascending=False)

print(user_5_ratings)

```

```python
# evaluating predictions using RMSE



# Extract the ground truth to compare your predictions against (first 20 rows and first 100 columns)
actual_values = user_ratings_table.iloc[:20, :100].values

predicted_values = calc_pred_ratings_df.iloc[:20, :100].values

# Create a mask of actual_values to only look at the non-missing values in the ground truth
mask = ~np.isnan(actual_values)

# Print the performance of the predictions

print(mean_squared_error(actual_values[mask], predicted_values[mask], squared=False))
```
