```python
results = page_soup.find_all(class_='description')

data = []

for result in results:
    
    title = result.h3.text
    text = result.find(class_='mob-hidden').text
    data.append((title, text))
```

# scraping for links to song lyrics

```python
base_url = 'http://www.lyrics.com/artist/madonna'


artist_page = requests.get(base_url)

print(artist_page.text)
```

```python
artist_soup = BeautifulSoup(artist_page.text)

artist_soup.find_all(class_='tal qx')
```

```python
links = artist_soup.find_all(class_='tal qx')
BASE_URL = 'http://www.lyrics.com'

urls = {}

for link in links:
    print(link.find('a'))
    print(link.find('a').get('href'))
    print(link.find('a').text)
    song_name = link.find('a').text
    song_url = link.find('a').get('href')
    url = BASE_URL+song_url
    urls[song_name] = url
    
```

```python
urls
```

```python
len(urls)
```

```python
# extract song lyrics

eg_url = 'https://www.lyrics.com/lyric/36050507/Madonna/Live+to+Tell'

song_html = requests.get(eg_url)

print(song_html.text)
```

```python
page_soup = BeautifulSoup(song_html.text)

page_soup
```

```python
lyrics = page_soup.find(class_='lyric-body').text

lyrics
```

```python
# clean text

lyrics = lyrics.lower()

lyrics = re.sub('\n', ' ', lyrics)  # removed the '\n' 

lyrics = re.sub('[^\w\s]', ' ', lyrics) # removed all punctuation characters

lyrics
```
