```python
import re
import scrapy
import requests
from scrapy import Selector, Request, Spider
from scrapy.crawler import CrawlerProcess
import pandas as pd
```

## 1. Scraping Lyrics

```python

class Lyrics_Spider(scrapy.Spider):
  name = "lyrics_spider"
  
  # start_requests method
  def start_requests(self):
    yield scrapy.Request(url = "https://genius.com/artists/Ed-sheeran", callback = self.parse_album_links)
  
  # First parsing method
  def parse_album_links(self, response):
    album_links = response.css('a.vertical_album_card::attr(href)').extract()
    print(album_links)
    albums.extend(album_links)
    for album_link in album_links:
        yield response.follow(url = album_link, callback = self.parse_links)
  
  # Second parsing method            
  def parse_links(self, response):
    lyrics_links = response.css('a.u-display_block::attr(href)').extract()
    links.extend(lyrics_links)
    for lyrics_link in lyrics_links:
      yield response.follow(url = lyrics_link, callback = self.parse_lyrics)
  
  # Third parsing method
  def parse_lyrics(self, response):
    song_title = response.css('title::text').extract_first().strip()
    song_lyrics = response.xpath(
            ('/html//*[contains(@class,"lyric")]//a/text()|/html//*[contains(@class,"lyric")]//p/text()')).extract()
    
    song_lyrics = [line.strip() for line in song_lyrics]
    lyrics_dict[song_title] = song_lyrics
```

```python
# Initialize the dictionary **outside** of the Spider class
links = []
albums = []
lyrics_dict = dict()

# Run the Spider
process = CrawlerProcess()
process.crawl(Lyrics_Spider)
process.start()
```

## 2. Cleaning Lyrics

```python
print(lyrics_dict.keys())
```

```python
lyrics_clean = lyrics_dict.copy()
new_keys = []
# cleaning keys to show only song name, marking remixes as remix to drop later
for key in lyrics_clean.keys():
    parts = key.replace("|", "–").split("–")
    new_key = parts[1].lower().replace("lyrics", "").strip()
    new_key = re.sub(r"\[.*?\]", "", new_key)
    new_key = re.sub(r"\(.*?\)", "remix", new_key)
    new_key = re.sub(r"\W+", " ", new_key)
    print(new_key)
    new_keys.append(new_key)

#print(lyrics_clean.keys())
```

```python
print(len(lyrics_clean.keys()), len(new_keys))
```

```python
# replace all keys with new cleaned keys
lyrics_clean = dict(zip(new_keys, list(lyrics_clean.values())))
lyrics_clean.keys()
```

```python
# step 1: join lyrics into one string
clean = " ".join(exp1)
```

```python
# step 2: lower case all words in lyrics and strip white spaces in the beginning and end

clean = clean.lower().strip()

#print(lyrics_clean)
```

```python
# step 3: # remove anything inside brackets
brackets_regex = r"\[.*?\]|\(.*?\)|\u2005"
print(re.findall(brackets_regex, clean))
```

```python
clean = re.sub(brackets_regex, "", clean)
clean
```

```python
# step4: remove punctuation and double spaces
clean = re.sub(r"\W+", " ", clean)
clean = clean.strip()
clean
```

```python
# cleaning up titles and lyrics and saving in text files
lyrics_dict_copy = lyrics_dict.copy()
clean_titles = []
clean_lyricss = []


for title, lyrics in lyrics_dict_copy.items():
    # clean titles first
    parts = title.replace("|", "–").split("–")
    clean_title = parts[1].lower().replace("lyrics", "").strip()
    clean_title = re.sub(r"\[.*?\]", "", clean_title)
    clean_title = re.sub(r"\(.*?\)", "remix", clean_title)
    clean_title = re.sub(r"'+", "", clean_title)
    clean_title = re.sub(r"\W+", " ", clean_title)
    clean_title = clean_title.strip()
    clean_titles.append(clean_title)
    
    #clean lyrics
    clean_lyrics = " ".join(lyrics)
    clean_lyrics = clean_lyrics.lower().strip()
    clean_lyrics = re.sub(r"\[.*?\]|\(.*?\)|\u2005", "", clean_lyrics)
    clean_lyrics = re.sub(r"'+", "", clean_lyrics)
    clean_lyrics = re.sub(r"\W+", " ", clean_lyrics)
    clean_lyrics = clean_lyrics.strip()
    clean_lyricss.append(clean_lyrics)
    
lyrics_dict_clean = dict(zip(clean_titles, clean_lyricss))
```
